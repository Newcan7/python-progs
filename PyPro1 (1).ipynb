```python
class PointClass:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
        return f"Point ({self.x}, {self.y})"
    def distance_to(self, other_point):
        return ((self.y - other_point.y)**2 + (self.x - other_point.x)**2)**0.5
    def distance_to_origin(self):
        return (self.y**2 + self.x**2)**0.5
    def lies_on_line(self, line):
        return (line.a*self.x + line.b*self.y + line.c) == 0

class LineClass:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def __str__(self):
        return f"Line {self.a}x + {self.b}y + {self.c} = 0"
    def distance_from_point(self, point):
        return abs((self.a*point.x + self.b*point.y)/((self.a**2 + self.b**2)**0.5))
    def distance_from_line(self, line):
        return abs((self.c - line.c)/abs(((self.a**2 + self.b**2)**0.5)))

#From Outside:
lc1 = LineClass(6, 1, 0)
lc2 = LineClass(-6, -1, 0)
pc1 = PointClass(-1, 6)
pc2 = PointClass(1, -6)

print(lc1)
print("Distance from", pc1, "is :", lc1.distance_from_point(pc1))
print("Distance from", lc1, "to", lc2, ":", lc1.distance_from_line(lc2))
print("\n")
print(pc1)
print("Distance from", pc1, "to", pc2, ":", pc1.distance_to(pc2))
print("Distance from", pc1, "to Origin :", pc1.distance_to_origin())
print("Does", pc1, "lie on", lc1, ":", pc1.lies_on_line(lc1))
```

```python
class OrderSystem:
    order_counter = 1 #Static variable
    def __init__(self, customer_name):
        self.order_id = OrderSystem.order_counter
        OrderSystem.order_counter = OrderSystem.order_counter + 1
        self.customer_name = customer_name
        self.cart_total = 0
        self.__discount_percent = 10 ##private
        self.final_amount = 0
        self.paid = False
    
    def add_to_cart(self, amount):
        self.cart_total = self.cart_total + amount
        return f"Items Added! Cart Total Rs. {(self.cart_total)}"
    
    def apply_discount(self):
        discount_amount = (self.cart_total * self.__discount_percent) / 100
        self.final_amount = self.cart_total - discount_amount
        return f"Discount Applied! You saved Rs. {discount_amount}"
    
    def process_payment(self):
        if self.final_amount == 0:
            self.apply_discount()
        if not self.paid: 
            self.paid = True
            return f"Payment of Rs. {self.final_amount} is successful!"
        return "Already Paid!"
        
    def show_details(self):
        return f"""
        Order ID: {self.order_id}
        Customer Name: {self.customer_name}
        Cart Total: {self.cart_total}
        
        Discount Percent: {self.__discount_percent}
        Final Amount: {self.final_amount}
        Payment Status: {"PAID" if self.paid else "PAYMENT PENDING"}
        """

ob1 = OrderSystem("Anurag-San")
#ob1.__discount_percent = 70 ## Of course, you can attempt to 'modify' it but it won't reflect in the original definition 
#ob1._OrderSystem__discount_percent = 70 ## Surprise MF!!! Of course you can modify it at the class level! This is called NAME MANGLING

print(ob1.add_to_cart(1200), "\n", ob1.apply_discount(), "\n", ob1.process_payment(), "\n", ob1.show_details())
```

```python
#Use keyword 'super' to prevent parent class's methods from being overridden by child class. 'super' only works for methods.
class Parent:
    def __init__(self, x, y):
        print("Hello from parent.")
        self.__x = x
        self.y = y
    def call_method(self):
        print("Calling from parent!")

class Child(Parent):
    def call_method(self):
        super().call_method()
        print("Calling from child!")

ob1 = Child(100, 20)
ob1.call_method()
```

```python
class Fraction:
    def __init__(self, num, deno):
        self.num = num
        self.deno = deno

    def __str__(self):
        return f"{self.num} / {self.deno}"

    def add(self, x):
        return f"{self.num * x.deno + x.num * self.deno} / {self.deno * x.deno}"

    def sub(self, x):
        return f"{self.num * x.deno - x.num * self.deno} / {self.deno * x.deno}"

    def mult(self, x):
        return f"{self.num * x.num} / {self.deno * x.deno}"

    def div(self, x):
        return f"{self.num  / self.deno} / {x.num / x.deno}"

ob1 = Fraction(12, 6)
ob2 = Fraction(12, 6)
dct = {'Add' : ob1.add(ob2), 'Sub' : ob1.sub(ob2), 'Mult' : ob1.mult(ob2), 'Div' : ob1.div(ob2)}
dct
```

```python
# We can make indexes are strings too!

r_Series = pd.Series([20,21,12], index = ['LONDON', 'NEW YORK', 'HELSINKI'])
print(r_Series)

#Series can use apply to use any function
def sq(x):
    return x**2
print(r_Series.apply(sq))

# Or you may use Lambda Functions as well!
print(r_Series.apply(lambda lm: lm**2))

def sub_Value(x, value, v1):
    return x- value + v1
r_Series.apply(sub_Value, args = [5, 6]) # x = 5, value = 6, v1 = r_Series values!
```

```python
# Create a series containing 40 random elements between 1-100; If elements >= 50 replace with HIGH else LOW. Count such LOWs and HIGHs.

r_Series = pd.Series(np.random.randint(1,100, size = 40))
r_Series = r_Series.apply(lambda lm: 'High' if lm >= 50 else 'Low')
print(r_Series)
print("No. of HIGHs : ", r_Series[r_Series == 'High'].size,"\nNo. of LOWs : ",r_Series[r_Series == 'Low'].size)
```

```python
#Keyword Arguments

s = pd.Series(np.random.randint(1,10, size = 5))
print(s)
def add_custom_values(x, **kwargs):
    for month in kwargs:
        x += kwargs[month]
    return x

s.apply(add_custom_values, june = 30, july = 20, august = 25) #Variable arguments being added!
```
