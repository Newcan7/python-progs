```python
a=int(input("Enter the First number:"))
b=int(input("Enter the Second number:"))
c=input("Enter the operation you want to do:")
match c:
    case "+":
        print("result of addition is:", a+b)
    case "-":
        print("result of subtraction is:", a-b)
    case "*":
        print("result of multiplication is:", a*b)
    case "/":
        print("result of division is:", a/b)
    case "%":
        print("result of mod is:", a%b)
    case "and":
        print("result of and is:", a and b)
    case "or":
        print("result of or is:", a or b)
    case "not":
        print("result of not is:", not a ,not b)
    case "&":
        print("result of bitwise and is:", a & b)
    case "|":
        print("result of bitwise or is:", a | b)
    case "==":
        print("result of equal to is:", a == b)
    case "!=":
        print("result of not equal to is:", a != b)
    case "<=":
        print("result of greater than equal to is:", a <= b)
    case ">=":
        print("result of lesser than equal to is:", a >= b)
    case _:
        print("Operation not possible:")

```
