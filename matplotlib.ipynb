```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
plt.xlabel('height')
plt.ylabel('weight')
plt.title('Graph', fontsize=15)
plt.plot(x1, y1, x2, y2)
plt.grid(axis='x')
plt.show()
```

```python
roll = np.array([0, 1, 2, 3])
#plot1
phy = np.array([83, 88, 91, 80])
plt.subplot(2, 2, 1)
plt.plot(roll, phy, 'bo', ms=10)
plt.title('Physics')

#plot2
chem = np.array([70, 76, 83, 75])
plt.subplot(2, 2, 2)
plt.plot(roll, chem, 'r+', ms=10)
plt.title('Chemistry')

#plot3
maths = np.array([88, 70, 91, 62])
plt.subplot(2, 2, 3)
plt.plot(roll, maths, 'y*', ms=10)
plt.title('Maths')

#plot4
bio = np.array([60, 67, 94, 77])
plt.subplot(2, 2, 4)
plt.plot(roll, bio, 'm^', ms=10)
plt.title('Biology')

plt.suptitle('Marks Distribution')
plt.show()
```

```python
x = np.array([5, 15, 35, 25, 10])
y = np.array([100, 110, 120, 90, 80])

plt.scatter(x, y)
plt.show()
```

```python
x = np.array([5, 15, 35, 25, 10])
y = np.array([100, 110, 120, 90, 80])
colours = np.array([0, 20, 40, 60, 80])
sizes = np.array([200, 300, 400, 500, 600])
plt.scatter(x, y, s=sizes, c=colours, cmap='jet', alpha=0.6)
plt.colorbar()
plt.show()
```
