```python
plt.plot(sales.facecream,marker='o',linewidth=3)
plt.plot(sales.facewash,marker='o',linewidth=3)
plt.plot(sales.toothpaste,marker='o',linewidth=3)
plt.plot(sales.bathingsoap,marker='o',linewidth=3)
plt.plot(sales.shampoo,marker='o',linewidth=3)
plt.plot(sales.moisturizer,marker='o',linewidth=3)
plt.legend(loc='upper left',labels=["facecream sales data","facewash sales data","toothpaste sales data","bathingsoap sales data","shampoo sales data","moisturizer sales data"])
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('sales data')
plt.xticks(x)

plt.show()
```

```python
plt.plot(sales.toothpaste,"o")
plt.xlabel('Month Number')
plt.ylabel('number of units sold')
plt.title('toothpaste sales data')
plt.grid(linestyle='--')
plt.legend(loc='upper left',labels=['toothpaste sales data'])
plt.xticks(x)

plt.show()
```

```python
x=np.array(sales.month_number)
plt.bar(x-0.15,sales.facecream,width=0.3)
plt.bar(x+0.15,sales.facewash,width=0.3)
plt.legend(loc='upper left',labels=['facecream sales data','facewash sales data'])
plt.grid(linestyle="--")
plt.title('facewash and facecream sales data')
plt.ylabel('sales units in numbers')
plt.xlabel('month number')
plt.xticks(x)
plt.show()
```

```python
plt.bar(sales.month_number,sales.bathingsoap)
plt.grid(linestyle='--')
plt.title('bathingsoap sales data')
plt.ylabel('sales units in numbers')
plt.xlabel('month number')
plt.xticks(x)
plt.show()
```

```python
values=[sales.facecream.sum(),sales.facewash.sum(),sales.toothpaste.sum(),sales.bathingsoap.sum(),sales.shampoo.sum(),sales.moisturizer.sum()]
labels=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
plt.pie(values,labels=labels,autopct='%1.1f%%')
plt.title('sales data')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(sales.bathingsoap,'-k',marker='o',linewidth=3)
plt.title('sales data of bathingsoap')
plt.xticks(x)

plt.subplot(2,1,2)
plt.plot(sales.facewash,'-r',marker='o',linewidth=3)
plt.title('sales data of facewash')
plt.xlabel('month number')
plt.ylabel('sales units in numbers')
plt.xticks(x)

plt.show()
```
