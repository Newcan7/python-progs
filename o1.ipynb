```python
a=int(input("Enter the first number"))
b=int(input("Enter the second number"))
c=input("Enter the operation to be performed with the two numbers")
match c:
    case '+':
        print(a+b)
    case '-':
        print(a-b)
    case '*':
        print(a-b)
    case '/':
        print(a-b)
    case '%':
        print(a-b)
    case '**':
        print(a-b)
    case '//':
        print(a-b)
    case _:
        print("Wrong choice")
```
