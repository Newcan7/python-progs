```python
s=input("Enter password: ")
l=len(s)
lw=u=sc=r=0
for i in s:
    if i.islower()==True:
        lw=1
    elif i.isupper()==True:
        u=1
    elif i.isdigit()==True:
        d=1
    else:
        sc=1

        
if l>=8 and lw==1 and u==1 and d==1 and sc==1:
    print("strong")
elif l<=8 or lw==0 or u==0 or d==0 or sc==0:
    print("moderate")
else:
    print("weak")
```
