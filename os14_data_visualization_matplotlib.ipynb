```python
x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])
x2=np.array([0,1,2,3])
y2=np.array([6,2,7,11])
plt.xlabel('height')
plt.ylabel('weight')
plt.title('graph',fontsize=15)

plt.plot(x1,y1,x2,y2)
plt.grid()
# plt.grid(axis='x')
plt.show()
```

```python
# plot 1
roll=np.array([0,1,2,3])
phy=np.array([83,88,91,80])

plt.subplot(2,2,1) # row, column, plot number
plt.plot(roll,phy,'bo',ms=10)
plt.title('physics')

# plot 2
chem=np.array([70,76,83,75])

plt.subplot(2,2,2) # row, column, plot number
plt.plot(roll,chem,'r+',ms=10)
plt.title('chemistry')

# plot 3
maths=np.array([65,80,75,70])

plt.subplot(2,2,3) # row, column, plot number
plt.plot(roll,maths,'g^',ms=10)
plt.title('maths')

# plot 4
bio=np.array([89,90,79,85])

plt.subplot(2,2,4) # row, column, plot number
plt.plot(roll,bio,'c*',ms=10)
plt.title('biology')

plt.suptitle('marks distribution')
plt.show()
```

```python
xx=np.array([5,15,35,25,10])
yy=np.array([100,110,120,90,80])
plt.scatter(xx,yy)
# used to show the distribution of data in a graph
plt.show()
```

```python
xx=np.array([5,15,35,25,10])
yy=np.array([100,110,120,90,80])
colours=np.array([0,20,40,60,80])
sizes=np.array([200,300,400,500,600])
plt.scatter(xx,yy,s=sizes,c=colours,cmap='spring',alpha=0.6)
# spring summer winter autumn
# alpha specifies the opacity of the markers. if not mentioned then 1 or completely opaque
plt.colorbar()
plt.show()
```

```python
months=np.array(['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','now','dec'])
r1=np.array([40,50,60,70,80,85,75,80,75,60,55,40])
r2=np.array([30,40,50,60,70,75,65,70,65,50,45,30])
plt.scatter(months,r1,cmap='spring')
plt.scatter(months,r2,cmap='spring')
plt.colorbar()
plt.show()
```
