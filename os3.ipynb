```python
def factorial(n):
    fac=1
    for i in range (2,n+1):
        fac=fac*i
    return fac
factorial(int(input("Enter number")))
```

```python
def prime(n):
    fac=0
    for i in range (2,int(n/2)):
        if n%i==0:
            fac=fac+1
    if fac==0:
        return True
    else:
        return False
prime(int(input("Enter number")))
```

```python
def prime(n):
    ls=[]
    for i in range (1,n+1):
        if n%i==0:
            fcs=0
            for j in range (1,i+1):
                if i%j==0:
                    fcs=fcs+1
            if fcs==2:
                ls.append(i)
    return ls

prime(int(input("Enter number")))
```

```python
def pf(n):
    fc=[]
    for i in range(2,n+1):
        while n%i==0:
            fc.append(i)
            n//=i
    return fc
pf(10)
```
