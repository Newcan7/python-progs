```python
l=[1,2,3,4,5,6,7,8,9]
def prime(n):
    f=0
    for i in range(1,n+1):
        if n%i==0:
            f=f+1
    if f==2:
        return True
    else:
        return False
t=filter(lambda x:prime(x)==False,l)
print(list(t))
```
