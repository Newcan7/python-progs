```python
class Point: 
    def __init__(self,x,y): 
        self.x=x 
        self.y=y 
    def __str__(self): 
        return f"({self.x},{self.y})" 
    def distance_to(self,o): 
        d=(((self.x-o.x)**2)+((self.y-o.y))**2)**0.5 
        return d 
    def distance_to_origin(self):
        ds=((self.x)**2+(self.y)**2)**0.5
        return ds
obj=Point(2,3) 
obj2=Point(1,2) 
print(obj) 
print(obj2) 
print(obj.distance_to(obj2))
print(obj.distance_to_origin())
```

```python
class Line:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def __str__(self):
        return f"{self.a}x + {self.b}y + {self.c}"
    def distance_from_point(self,o):
        return abs(self.a*o.x+self.b*o.y+self.c)/((self.a**2+self.b**2)**0.5)
    def distance_from_line(self,l):
        return abs(self.c-l.c)/((self.a**2+self.b**2)**0.5)
ob=Line(1,2,3)
print("line 1: ",ob)
o=Point(1,2)
print("point: ",o)
print("distance from point to line 1: ",ob.distance_from_point(o))
o1=Line(1,2,4)
print("line 2: ",o1)
print("distance from line 1 to line 2: ",ob.distance_from_line(o1))
```
