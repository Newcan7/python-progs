```python
class OrderSystem:
    __oc=1
    def __init__(self,cnm):
        self.__oid=OrderSystem.__oc #order id
        OrderSystem.__oc=OrderSystem.__oc+1
        self.__cnm=cnm #customer name
        self.__ctl=0 #cart total
        self.__dcp=10 #discount percent
        self.__fam=0 #final amount
        self.__paid=False

    @staticmethod
    def get_order_if():
        return OrderSystem.__oc

    def get_discount(self,percent): #incapsulation, getter method
        self.__dcp=percent

    def add_to_cart(self,amount):
        self.__ctl=self.__ctl+amount
        return f"items added! cart total Rs.{self.__ctl}"
        
    def apply_discount(self):
        __dam=self.__ctl*self.__dcp/100 #discount amount
        self.__fam=self.__ctl-__dam
        
        return f"discount applied! you save Rs.{__dam}"
    def process_payment(self):
        if self.__fam==0:
            self.apply_discount()
        if not self.__paid:
            self.__paid=True
            return f"payment of Rs.{self.__fam} is successful"
        return "already paid"
        
    def show_details(self):
        return f"""
        order id: {self.__oid}
        customer name: {self.__cnm}
        cart total:{self.__ctl}
        discount percent: {self.__dcp}
        final amount: {self.__fam}
        payment status: {"paid" if self.__paid else "pending"}
        """
OrderSystem.records=[]
obj=OrderSystem("s")

obj.__dcp=90
print(obj.__dcp)
print(obj._OrderSystem__dcp)
obj._OrderSystem__dcp=70 #does work but illegal

obj.get_discount(10) 

print(obj.add_to_cart(5000))
print(obj.apply_discount())
print(obj.process_payment())
print(obj.show_details())
records.append(obj.show_details())
print(records)
```

```python
class parent:
    def __init__(self,x):
        self.x=x
        print(f"this is x: {self.x} from parent")
    def method_parent(self):
        print("hello from parent")
class child(parent):
    def __init__(self,y): #method overriding
        self.y=y
        print(f"this is y: {self.y} from child")
    def method_child(self):
        print("hello from child")
obj1=child(7)
obj1.method_parent()
obj1.method_child()
```
