```python
class parent:
    def __init__(self,x,y):
        print("hello from parent")
        self.__x=x
        self.y=y
    def get_x(self):
        print(self.__x)
class child(parent):
    def test_private(self):
        print(f"y: {self.y}") #public attributes can be inherited
        print(f"x: {self.__x}") #private attributes cannot be inherited
        
obj=child(100,20)
obj.test_private()
```

```python
class parent:
    def __init__(self,x,y): #constructor has to run since child class doesnt have a contructor and parent class does
        print("hello from parent")
        self.__x=x
        self.y=y
    def call_method(self):
        print("calling from parent")
class child(parent):
    def call_method(self):
        super().call_method()
        print("calling from child")
        
obj1=child(100,20)
obj1.call_method()
```

```python
class parent:
    def __init__(self,x,y): 
        print("hello from parent")
        self.__x=x
        self.y=y
    def call_method(self):
        print("calling from parent")
class child(parent):
    def __init__(self,p,q):
        self.p=p
        self.q=q
        super().__init__(p,q)
    def call_method(self):
        print(super().y) #super cannot access data/variables, super only works for methods
        print("calling from child")
        
obj1=child(100,20)
obj1.call_method()
```

```python
from abc import ABC, abstractmethod #abc is a module
#abstract class
class vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
    @abstractmethod
    def stop(self):
        pass

class car(vehicle):
    def start(self):
        print("car: turn key -> engine starts")
    def stop(self):
        print("car: press brake -> engine stop")

class motorcycle(vehicle):
    def start(self):
        print("bike: kickstart -> engine starts")
    def stop(self):
        print("bike: kick brake -> engine stop")

vehicles=[car(),motorcycle()]
print("starting all vehicles")
for vehicle in vehicles:
    vehicle.start()

print("stopping all vehicles")
for vehicle in vehicles:
    vehicle.stop()
    
```

```python
class fraction:
    def __init__(self,d,n):
        self.d=d
        self.n=n
    def __str__(self):
        return f"{self.d}/{self.n}"
    def add(self,o):
        nu=int((self.n*o.d+o.n*self.d))
        de=int((self.d*o.d))
        return f"{nu}/{de}"
obj=fraction(5,8)
obj1=fraction(6,4)
print(obj)
print(obj1)
obj.add(obj1)
```
