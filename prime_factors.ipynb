```python
def prime(a):
    c=2
    flag=True
    for c in range(2,int(a/c)+1):
        if(a%c==0):
            flag=False
            break
    return flag

def factors(n):
    l=[]
    for i in range (1,n):
        if n%i==0:
            if prime(i)==True:
                l.append(i)
    return l
k=int(input("Enter a number"))
print(factors(k))
```
