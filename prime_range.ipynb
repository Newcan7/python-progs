```python
def prime(a):
    c=2
    flag=True
    for c in range(2,int(a/c)+1):
        if(a%c==0):
            flag=False
            break
    return flag
    
n = int(input('enter limit'))
for i in range(1,n+1):
    if(prime(i)==True):
        print('prime number detected : ',i)
    else:
        continue
```
